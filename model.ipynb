```python
import numpy as np
import pandas as pd
from sklearn import metrics
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import BaggingClassifier,RandomForestClassifier
from sklearn.naive_bayes import MultinomialNB,BernoulliNB,ComplementNB,GaussianNB
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, roc_auc_score,classification_report
from sklearn.model_selection import GridSearchCV,RandomizedSearchCV
from sklearn.multioutput import MultiOutputClassifier
```

```python
df=pd.read_csv('train.csv')
```

```python
df
```

```python
x=df.drop(['respondent_id'],axis=1)
```

```python
x
```

```python
df2=pd.read_csv('training_set_labels.csv')
```

```python
df2
```

```python
y=df2[['xyz_vaccine','seasonal_vaccine']]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python

model = RandomForestClassifier(
    bootstrap=True,
    max_depth=20,
    min_samples_leaf=4,
    min_samples_split=10,
    n_estimators=300
)
multi_target_rf = MultiOutputClassifier(model, n_jobs=-1)

multi_target_rf.fit(x_train, y_train)

y_prob = multi_target_rf.predict_proba(x_test)

prob_df = pd.DataFrame({
    'output1_prob_0': y_prob[0][:, 0],  
    'output1_prob_1': y_prob[0][:, 1],  
    'output2_prob_0': y_prob[1][:, 0],  
    'output2_prob_1': y_prob[1][:, 1]   
})


y_pred = multi_target_rf.predict(x_test)

print("Classification Report for first output:")
print(classification_report(y_test.iloc[:, 0], y_pred[:, 0]))
print("Accuracy for first output:", accuracy_score(y_test.iloc[:, 0], y_pred[:, 0]))

print("\nClassification Report for second output:")
print(classification_report(y_test.iloc[:, 1], y_pred[:, 1]))
print("Accuracy for second output:", accuracy_score(y_test.iloc[:, 1], y_pred[:, 1]))
```

```python

print("roc_auc for first output:", roc_auc_score(y_test.iloc[:, 0], y_pred[:, 0]))

print("roc_auc for second output:", roc_auc_score(y_test.iloc[:, 1], y_pred[:, 1]))
```
